<div style="text-align:center;"><h1>Generating a new Directed Network</h1></div>

<h2>Dependencies</h2>

```python
%load_ext autoreload
%autoreload 2
import warnings

warnings.filterwarnings("ignore")
from org.gesis.model import network
from org.gesis.model.network import DirectedNetwork
from org.gesis.libs import viz
```

<h2>Parameters</h2>

```python
kind = network.kDH           # kind: kDBA, kDH, kT, kHDBA, kHTDBA
N = 100                      # number of nodes
m = 1                        # minimum degree
density = 0.01               # edge density
minority_fraction = 0.1      # minority fraction
gamma_m = 3                  # exponent degree power-law minority 
gamma_M = 3                  # exponent degree power-law majority
h_mm = 1.0                   # homophily between minorities
h_MM = 1.0                   # homophily between majorities
triads_ratio = None          # triads density
triads_pdf = [1/12]*12       # pdf for 12 triads: [mmm, mnm, MMM, MNM, mmN, mmM, mMn, nMm, nMM, mMM, mNM, mMN]
```

<h2>Network generation</h2>

```python
DN = DirectedNetwork(kind,
                     N=N,
                     m=m,
                     density=density,
                     minority_fraction=minority_fraction,
                     gamma_m=gamma_m,
                     gamma_M=gamma_M,
                     h_mm=h_mm,
                     h_MM=h_MM,
                     triads_ratio=triads_ratio,
                     triads_pdf=triads_pdf)
DN.create_network()
DN.info()
```

<h2>Plots</h2>

```python
viz.plot_simple_network(DN.G)
```
