<h2>Loading metadata distributions</h2>

```python
df_metadata = network.load_all_node_metadata_synthetic(MODEL, OUTPUT)
```

```python
df_metadata.sample(5)
```

```python
df_metadata_sym = df_metadata.query("hmm==hMM").copy()
hs = [0.0, 0.2, 0.5, 0.8, 1.0]
df_metadata_asym = df_metadata.query("hmm in @hs and hMM in @hs").copy()
```

<h1>Symmetric Homophily</h1>

<h2>In-degree distribution</h2>

```python
metric = 'indegree'
```

```python
fn = os.path.join(PLOTS, '{}_synthetic_vertical_inequality_symmetric_homophily.pdf'.format(metric))
viz.plot_lorenz_curve(df_metadata_sym, metric, col='hmm', row='fm', fn=fn)
```

```python
fn = os.path.join(PLOTS, '{}_synthetic_vertical_inequality_symmetric_homophily_per_group.pdf'.format(metric))
viz.plot_lorenz_curve(df_metadata_sym, metric, groups=True, col='hmm', row='fm', fn=fn)
```
